kmeans是一个class,通过对变量赋值为kmeans对象,并使用对象内置函数以实现k means cluster算法。

注意事项：

本对象所用数据结构均基于pandas.DataFrame和pandas.Series构造，因此要求kmeans.fit(data)和kmeans.dis_compute(sample,center)中的data, sample, center为此两类结构。

本程序目前仅考虑使用欧氏距离作为样本点之间的距离，其余距离表达形式暂未考虑。

本程序采用所有样本到簇中心距离平方和作为模型拟合好坏的评价标准，公式如下：

其中k为簇个数，n为样本个数。

对象参数：

k：想要聚类簇的个数

max_iter：算法最大迭代次数，默认为500次

threshold：两次所有样本到簇中心距离平凡和的差值小于threshold，即认为继续迭代不会对算法有显著提成，从而停止算法，默认为0.01

对象特质：

kmeans.k=k

kmeans.max_iter=max_iter

kmeans.threshold=threshold

kmeans.center：      输出所有簇的中心点

kmeans.label：        输出所有样本所在的簇标签，以范围 [1,k] 之间的整数表示标签

kmeans.label_dis：    输出每个样本与所在簇中心的距离

kmeans.score：        输出所有样本到簇中心距离平方和

对象函数：

kmeans.dis_compute(sample,center)： 计算单个样本sample到所有中心center的距离，输出sample所在簇的标签，以及与所在簇的距离

kmeans.fit(data)： 对data进行k means cluster算法                  

```python
import pandas as pd
import numpy as np
import random
```

```python
class kmeans:

    def __init__(self,k,max_iter=500,threshold=0.01):
        #k是聚类簇的个数
        #max_iter是算法最大的迭代次数
        #threshold是算法收敛的条件，评价系数小于threshold即停止算法
        self.k=k
        self.max_iter=max_iter
        self.threshold=threshold

    
    def dis_compute(self,sample,center):
        #dis_compute计算每个样本点到所有中心的距离
        #返回样本点所在簇，以及与簇中心的距离
        sample1=sample.as_matrix()
        center1=center.as_matrix()
        dis=pd.DataFrame((center1-sample1)*(center1-sample1))
        dis=dis.apply(axis=1,func=np.sum)
        label=np.argmax(dis)+1
        dis_label=np.max(dis)
        return [label,dis_label]
        
            
    def fit(self,data):
        #fit为kmeans算法实现
        #默认data是一个pandas.DataFrame文件   
        
        #随机选取k个样本作为初始中心
        center_row=random.sample(range(0,data.shape[0]),self.k) 
        center=data.iloc[center_row,:]      
        
        #迭代k means算法
        result_old=[]
        result=[]
        score_old=None
        
        for i in range(0,self.max_iter):
            
            if i >= 1:
                result_old=result
                
            #计算每个样本点到所有中心点的距离,并为其分类
            result=data.apply(axis=1,func=self.dis_compute,center=center)
            
            #计算评价系数
            if i >= 1:
                score_old=np.sum(result_old[1]*result_old[1])
                score_new=np.sum(result[1]*result[1])
                self.score=score_new
                #判断是否小于threshold并停止迭代 
                if  score_new - score_old < self.threshold:
                    break
            
            #获得新的中心点
            for j in range(0,self.k):
                center.iloc[j,:]=data[result[0]==j+1].apply(axis=0,func=np.mean)
            
            self.center=center
            self.label=result[0]
            self.label_dis=result[1]
```

Performance of kmeans:

```python
import matplotlib.pyplot as plt

#randomize data in two dimensions
x11=np.random.normal(0, 4, 500 )
x21=np.random.normal(0, 4, 500)
x12=np.random.normal(10, 4, 500)
x22=np.random.normal(10, 4, 500)
data=pd.DataFrame([x11,x21]).T.append(pd.DataFrame([x12,x22]).T)
plt.scatter(data[0:500][0],data[0:500][1],alpha=0.4,label=1)
plt.scatter(data[500:][0],data[500:][1],alpha=0.4,label=2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Positions of Samples')
plt.show()
```

```python
#comparation of kmeans and sklearn.cluster.KMeans
from sklearn.cluster import KMeans
#kmeans
A=kmeans(k=2,max_iter=100,threshold=0.0001)
A.fit(data)
#output centers of kmeans
print(A.center)

#sklearn.cluster.KMeans
B=KMeans(2,max_iter=100,tol=0.0001)
B.fit(data)
#output centers of sklearn.cluster.KMeans
print('\n',B.cluster_centers_)
```

```python
#display the performances
fig=plt.figure(figsize=[6,10])
ax1=fig.add_subplot(2,1,1)
ax1.scatter(data[A.label==1][0],data[A.label==1][1],label=1,alpha=0.4)
ax1.scatter(data[A.label==2][0],data[A.label==2][1],label=2,alpha=0.4)
ax1.set_xlabel('X1')
ax1.set_ylabel('X2')
ax1.set_title('Performance of kmeans')
ax2=fig.add_subplot(2,1,2)
ax2.scatter(data[B.labels_ ==0][0],data[B.labels_ ==0][1],label=1,alpha=0.4)
ax2.scatter(data[B.labels_ ==1][0],data[B.labels_ ==1][1],label=2,alpha=0.4)
ax2.set_xlabel('X1')
ax2.set_ylabel('X2')
ax2.set_title('Performance of sklearn.cluster.KMeans')
plt.show()
```
